```python
import os, json, shutil
from pathlib import Path

# ── Config ────────────────────────────────────────────────────────────────────
DS1 = "/content/person.v1i.coco"
DS2 = "/content/ya_ilehi_7elha_fi_wjouhna.v1i.coco"
OUT = "/content/merged_dataset"

# Final classes
CLASSES = ["person", "infotainment"]  # id=0, id=1

# Class name → final class id
CLASS_MAP = {
    "per": 0,
    "car-screeen": 1,
    "IC-screen": 1,
    "carscreen": 1,
}

def merge_split(sources, out_dir):
    """
    sources: list of (img_dir, ann_file) tuples to merge
    """
    os.makedirs(out_dir, exist_ok=True)

    merged = {
        "info": {},
        "licenses": [],
        "categories": [{"id": i, "name": n} for i, n in enumerate(CLASSES)],
        "images": [],
        "annotations": [],
    }

    img_id = 1
    ann_id = 1

    for img_dir, ann_file in sources:
        if not os.path.exists(ann_file):
            print(f"Skipping missing: {ann_file}")
            continue

        with open(ann_file) as f:
            data = json.load(f)

        # Build old category id → new class id map
        cat_remap = {}
        for cat in data["categories"]:
            name = cat["name"]
            if name in CLASS_MAP:
                cat_remap[cat["id"]] = CLASS_MAP[name]
            else:
                print(f"  WARNING: unknown class '{name}' — skipping")

        # Build old image id → new image id map
        img_id_map = {}
        for img in data["images"]:
            src = os.path.join(img_dir, img["file_name"])
            if not os.path.exists(src):
                continue
            # Copy image
            dst = os.path.join(out_dir, img["file_name"])
            if not os.path.exists(dst):
                shutil.copy2(src, dst)
            # Register image
            new_img = {**img, "id": img_id}
            merged["images"].append(new_img)
            img_id_map[img["id"]] = img_id
            img_id += 1

        # Remap annotations
        for ann in data["annotations"]:
            if ann["image_id"] not in img_id_map:
                continue
            if ann["category_id"] not in cat_remap:
                continue
            new_ann = {
                **ann,
                "id": ann_id,
                "image_id": img_id_map[ann["image_id"]],
                "category_id": cat_remap[ann["category_id"]],
            }
            merged["annotations"].append(new_ann)
            ann_id += 1

    # Save annotation file
    ann_out = os.path.join(out_dir, "_annotations.coco.json")
    with open(ann_out, "w") as f:
        json.dump(merged, f)

    print(f"✓ {out_dir}: {len(merged['images'])} images, {len(merged['annotations'])} annotations")
    return ann_out

# ── Merge each split ──────────────────────────────────────────────────────────

# TRAIN: both datasets have train
merge_split(
    sources=[
        (f"{DS1}/train", f"{DS1}/train/_annotations.coco.json"),
        (f"{DS2}/train", f"{DS2}/train/_annotations.coco.json"),
    ],
    out_dir=f"{OUT}/train"
)

# VALID: only DS2 has valid
merge_split(
    sources=[
        (f"{DS2}/valid", f"{DS2}/valid/_annotations.coco.json"),
    ],
    out_dir=f"{OUT}/valid"
)

# TEST: only DS2 has test
merge_split(
    sources=[
        (f"{DS2}/test", f"{DS2}/test/_annotations.coco.json"),
    ],
    out_dir=f"{OUT}/test"
)

print("\nDone! Merged dataset at:", OUT)
print("Classes:", CLASSES)
```

```python
import sys
sys.path.append("/content/D-FINE")

import torch
import torch.nn as nn
import torchvision.transforms as T
from PIL import Image, ImageDraw
import os

# ── Config ─────────────────
# ── Config ─────────────────
WEIGHTS = "/content/dfine_finetune_output/best_stg1.pth"
DFINE_ROOT = "/content/D-FINE"
CONFIG = f"{DFINE_ROOT}/configs/dfine/dfine_hgnetv2_m_finetune.yml"  # <--- use fine-tuned config
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

CLASSES = {0: "person", 1: "infotainment"}
COLORS  = {0: "red", 1: "blue"}
IMAGE_PATH = "/content/285689_Controler_l_interface_de_sa_voiture_d_un_simple_geste_de_la_main_le_futur_est_arrive.jpg"   # <-- change this
THRESHOLD = 0.5

# Our 4 finetuned classes (remapped to 0-indexed)
CLASSES = {0: "person", 1: "tv", 2: "laptop", 3: "cell phone"}
COLORS  = {0: "red",    1: "blue", 2: "green", 3: "orange"}

# ── Load model ────────────────────────────────────────────────────────────────
from src.core import YAMLConfig

cfg = YAMLConfig(CONFIG, resume=WEIGHTS)

if "HGNetv2" in cfg.yaml_cfg:
    cfg.yaml_cfg["HGNetv2"]["pretrained"] = False

checkpoint = torch.load(WEIGHTS, map_location="cpu")
state = checkpoint.get("ema", checkpoint).get("module", checkpoint.get("model", checkpoint))
cfg.model.load_state_dict(state)

class DFINEModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = cfg.model.deploy()
        self.postprocessor = cfg.postprocessor.deploy()

    def forward(self, images, orig_target_sizes):
        outputs = self.model(images)
        return self.postprocessor(outputs, orig_target_sizes)

model = DFINEModel().to(DEVICE).eval()
print(f"Model loaded on {DEVICE}")

# ── Preprocess ────────────────────────────────────────────────────────────────
transforms = T.Compose([
    T.Resize((640, 640)),
    T.ToTensor(),
])

def preprocess(image_path):
    image = Image.open(image_path).convert("RGB")
    orig_w, orig_h = image.size
    orig_size = torch.tensor([[orig_h, orig_w]], dtype=torch.float32).to(DEVICE)
    img_tensor = transforms(image).unsqueeze(0).to(DEVICE)
    return image, img_tensor, orig_size

# ── Inference ─────────────────────────────────────────────────────────────────
def run_inference(image_path, threshold=THRESHOLD):
    image, img_tensor, orig_size = preprocess(image_path)

    with torch.inference_mode():
        labels, boxes, scores = model(img_tensor, orig_size)

    labels = labels[0]
    boxes  = boxes[0]
    scores = scores[0]

    # Filter by threshold
    mask   = scores > threshold
    labels = labels[mask]
    boxes  = boxes[mask]
    scores = scores[mask]

    return image, labels, boxes, scores

# ── Draw results ──────────────────────────────────────────────────────────────
def draw_results(image, labels, boxes, scores):
    draw = ImageDraw.Draw(image)

    for label, box, score in zip(labels, boxes, scores):
        cls_id   = int(label)
        cls_name = CLASSES.get(cls_id, f"class_{cls_id}")
        color    = COLORS.get(cls_id, "white")
        x1, y1, x2, y2 = box.tolist()

        draw.rectangle([x1, y1, x2, y2], outline=color, width=3)
        draw.rectangle([x1, y1 - 16, x1 + len(cls_name) * 8 + 50, y1], fill=color)
        draw.text((x1 + 4, y1 - 15), f"{cls_name} {score:.2f}", fill="white")

    return image

# ── Run ───────────────────────────────────────────────────────────────────────
image, labels, boxes, scores = run_inference(IMAGE_PATH,THRESHOLD)

print(f"Detections: {len(scores)}")
for label, box, score in zip(labels, boxes, scores):
    cls_name = CLASSES.get(int(label), f"class_{int(label)}")
    print(f"  {cls_name}: {score:.3f} | box: {[round(v, 1) for v in box.tolist()]}")

result = draw_results(image, labels, boxes, scores)
result.save("/content/output.jpg")
print("Saved → /content/output.jpg")

# Display in Colab
from IPython.display import display
display(result)
```

```python
import sys, os, torch
import torch.nn as nn
import torchvision.transforms as T
from PIL import Image, ImageDraw
from pathlib import Path
from tqdm import tqdm
import json

# ── Paths ───────────────────────────────
DFINE_ROOT = "/content/D-FINE"
WEIGHTS = "/content/dfine_finetune_output/best_stg1.pth"
CONFIG = f"{DFINE_ROOT}/configs/dfine/dfine_hgnetv2_m_finetune.yml"
TEST_DIR = "/content/merged_dataset/test"
OUTPUT_DIR = "/content/test_results"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# ── Classes & colors ────────────────────
CLASSES = {0: "person", 1: "infotainment"}
COLORS = {0: "red", 1: "blue"}

# ── Add DFINE to path ──────────────────
sys.path.append(DFINE_ROOT)
from src.core import YAMLConfig

# ── Load model ─────────────────────────
cfg = YAMLConfig(CONFIG, resume=WEIGHTS)
if "HGNetv2" in cfg.yaml_cfg:
    cfg.yaml_cfg["HGNetv2"]["pretrained"] = False

checkpoint = torch.load(WEIGHTS, map_location="cpu")
state = checkpoint.get("ema", checkpoint).get("module", checkpoint.get("model", checkpoint))
cfg.model.load_state_dict(state)

class DFINEModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = cfg.model.deploy()
        self.postprocessor = cfg.postprocessor.deploy()

    def forward(self, images, orig_target_sizes):
        outputs = self.model(images)
        return self.postprocessor(outputs, orig_target_sizes)

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
model = DFINEModel().to(DEVICE).eval()
print(f"Model loaded on {DEVICE}")

# ── Preprocessing ──────────────────────
transforms = T.Compose([T.Resize((640, 640)), T.ToTensor()])

def preprocess(image_path):
    image = Image.open(image_path).convert("RGB")
    orig_w, orig_h = image.size
    orig_size = torch.tensor([[orig_h, orig_w]], dtype=torch.float32).to(DEVICE)
    img_tensor = transforms(image).unsqueeze(0).to(DEVICE)
    return image, img_tensor, orig_size

# ── Inference ─────────────────────────
def run_inference(image_path, threshold=0.5):
    image, img_tensor, orig_size = preprocess(image_path)
    with torch.inference_mode():
        labels, boxes, scores = model(img_tensor, orig_size)
    labels = labels[0]; boxes = boxes[0]; scores = scores[0]
    mask = scores > threshold
    labels = labels[mask]; boxes = boxes[mask]; scores = scores[mask]
    return image, labels, boxes, scores

# ── Draw results ───────────────────────
def draw_results(image, labels, boxes, scores):
    draw = ImageDraw.Draw(image)
    for label, box, score in zip(labels, boxes, scores):
        cls_id = int(label)
        cls_name = CLASSES.get(cls_id)
        color = COLORS.get(cls_id, "white")
        x1, y1, x2, y2 = box.tolist()
        draw.rectangle([x1, y1, x2, y2], outline=color, width=3)
        draw.rectangle([x1, y1-16, x1+len(cls_name)*8+50, y1], fill=color)
        draw.text((x1+4, y1-15), f"{cls_name} {score:.2f}", fill="white")
    return image

# ── IoU calculation ────────────────────
def compute_iou(box1, box2):
    # box = [x1, y1, x2, y2]
    x1 = max(box1[0], box2[0])
    y1 = max(box1[1], box2[1])
    x2 = min(box1[2], box2[2])
    y2 = min(box1[3], box2[3])
    inter_area = max(0, x2 - x1) * max(0, y2 - y1)
    box1_area = max(0, box1[2]-box1[0]) * max(0, box1[3]-box1[1])
    box2_area = max(0, box2[2]-box2[0]) * max(0, box2[3]-box2[1])
    union_area = box1_area + box2_area - inter_area
    if union_area == 0: return 0
    return inter_area / union_area

# ── Load test annotations ──────────────
with open("/content/merged_dataset/test/_annotations.coco.json") as f:
    test_ann = json.load(f)
fname2id = {img['file_name']: img['id'] for img in test_ann['images']}

# ── Run on test set ────────────────────
ious_per_image = {}
test_images = list(Path(TEST_DIR).glob("*.jpg")) + list(Path(TEST_DIR).glob("*.png"))

for img_path in tqdm(test_images, desc="Processing test images"):
    img_name = img_path.name
    image, labels, boxes, scores = run_inference(str(img_path), threshold=0.5)

    # Draw results
    result = draw_results(image, labels, boxes, scores)
    out_path = os.path.join(OUTPUT_DIR, img_name)
    result.save(out_path)

    # Compute IoU per GT box
    gt_ann = [ann for ann in test_ann["annotations"] if ann["image_id"] == fname2id[img_name]]
    pred_boxes = [box.tolist() for box in boxes]
    pred_labels = [int(l) for l in labels]
    ious = []
    for gt in gt_ann:
        gt_box = gt["bbox"]
        # convert [x, y, w, h] -> [x1, y1, x2, y2]
        gt_box = [gt_box[0], gt_box[1], gt_box[0]+gt_box[2], gt_box[1]+gt_box[3]]
        best_iou = 0
        for pb, pl in zip(pred_boxes, pred_labels):
            pb = [pb[0], pb[1], pb[2], pb[3]]
            if pl == gt["category_id"]:
                iou = compute_iou(pb, gt_box)
                if iou > best_iou:
                    best_iou = iou
        ious.append(best_iou)
    ious_per_image[img_name] = ious

# ── Print summary ──────────────────────
all_ious = [iou for ious in ious_per_image.values() for iou in ious]
print(f"✅ Done! Results saved in {OUTPUT_DIR}")
print(f"Average IoU over test set: {sum(all_ious)/len(all_ious):.3f}")
```
